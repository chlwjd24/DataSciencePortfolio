```python
import json
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import tensorflow_datasets as tfds
from tensorflow.keras import layers,models
from keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import Callback, EarlyStopping,ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
batch_size = 64
img_size = 224
fine_grain = False
num_classes = 50 if fine_grain else 8

root = Path('/content/public')
```

Some potentially useful functions for reading data from files.

```python
def load_target_names(path='public/info.json'):
    with open(path) as f:
        info = json.load(f)
    return info['fine'], info['coarse']

def load_image(file_path, image_size=112):
    img = tf.io.read_file(file_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = tf.image.resize(img, [image_size, image_size])
    return img

def create_dataset_tf(split, fine_grained, batch_size=32, image_size=112):
  df = pd.read_csv(root / f"{split}.csv")
  file_paths = df.filename.map(lambda fn: str(root / split / fn))
  labels = df.fine if fine_grained else df.coarse

  dataset = tf.data.Dataset.from_tensor_slices((file_paths, labels))
  dataset = dataset.map(lambda x, y: (load_image(x, image_size=image_size), y))
  dataset = dataset.batch(batch_size)
  dataset = dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
  return dataset


def create_dataset_sklearn(split, fine_grained, image_size=48, percent=0.1):
  dataset = create_dataset_tf(split=split, fine_grained=fine_grained, batch_size=1000, image_size=image_size)
  dataset = dataset.map(lambda x, y: (tf.image.rgb_to_grayscale(x), y))
  X = None
  Y = None
  n_batches = int(len(dataset) * percent)
  for x,y in dataset.take(n_batches):
    x, y = x.numpy(), y.numpy()
    x = x.reshape(x.shape[0], -1)
    X = x if X is None else np.vstack([X, x])
    Y = y if Y is None else np.hstack([Y, y])

  return X, Y
```

# Conventional ML Models

```python
# Extract the features from a pretrained model.
base_model = MobileNetV2(input_shape=(img_size, img_size, 3), include_top=False, weights='imagenet')
preprocess_input = preprocessing.Rescaling(scale=1./127.5, offset=-1)
inputs = layers.Input(shape=(img_size, img_size, 3))
x = preprocess_input(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
outputs = layers.Dense(128)(x)
feature_extractor = Model(inputs, outputs)
```

```python
train_features = feature_extractor.predict(train_ds)
val_features = feature_extractor.predict(val_ds)
test_features = feature_extractor.predict(test_ds)
```

```python
# Obtain labels for train, validation, and test sets
y_train = np.concatenate([y for _, y in train_ds], axis=0)
y_val = np.concatenate([y for _, y in val_ds], axis=0)
y_test = np.concatenate([y for _, y in test_ds], axis=0)

# Reshape feature vectors
X_train = train_features.reshape(train_features.shape[0], -1)
X_val = val_features.reshape(val_features.shape[0], -1)
X_test = test_features.reshape(test_features.shape[0], -1)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# K-Nearest Neighbors (KNN)
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_accuracy = knn.score(X_val, y_val)
print("KNN Accuracy:", knn_accuracy)
```

```python
# Random Forest (RF)
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_accuracy = rf.score(X_val, y_val)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
# Reduce dimensionality of feature vectors (using PCA)
from sklearn.decomposition import PCA

# Choose the desired number of principal components
n_components = 100

# Apply PCA to reduce the dimensionality of the feature vectors
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)

# Train SVM on the reduced feature vectors
svm = SVC()
svm.fit(X_train_pca, y_train)
svm_accuracy = svm.score(X_val_pca, y_val)
print("SVM Accuracy (with PCA):", svm_accuracy)
```

```python
# Multi-Layer Perceptron (MLP)
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
mlp_accuracy = mlp.score(X_val, y_val)
print("MLP Accuracy:", mlp_accuracy)
```

```python
# AdaBoost
adaboost = AdaBoostClassifier()
adaboost.fit(X_train, y_train)
adaboost_accuracy = adaboost.score(X_val, y_val)
print("AdaBoost Accuracy:", adaboost_accuracy)
```

## Notes on Conventional ML Models

By using a pretrained model by MobileNetV2, features were extracted first so that conventional models could be used. Then various models were tested against the public dataset. From the results above, SVM with PVA shows the highest accuracy score of 0.5689 and is therfore the best model out of K-nearest neighbours, random forests, multi-layer perception, adaboost and SVM. The kernel used with the support vector machine model is the radial basis function kernel.
The other models achieved an accuracy of 0.49248, 0.53584, 0.53546, and 0.47746 respectively.
The models performed were similarly ranked when performed on the fine grained dataset.

# Deep Learning Models


```python
keras.backend.clear_session()
```

```python
# Basic model
import tensorflow as tf
from tensorflow.keras import layers, models

# Define the model architecture
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_size, img_size, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Define callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=3, monitor='val_loss'),
    tf.keras.callbacks.ModelCheckpoint('best_model.h5', save_best_only=True),
    tf.keras.callbacks.ReduceLROnPlateau(factor=0.1, patience=2, min_lr=0.00001)
]

# Train the model
history = model.fit(train_ds,
                    steps_per_epoch=int(0.75 * len(train_ds) / batch_size),
                    validation_data=val_ds,
                    validation_steps=int(0.15 * len(train_ds) / batch_size), epochs=20, callbacks=callbacks)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(test_ds)
print('Test Loss:', test_loss)
print('Test Accuracy:', test_acc)
```

The simple CNN achieves an accuracy of around 

```python
# Xception model

base_model = keras.applications.xception.Xception(weights="imagenet",
                                                  include_top=False)
avg = keras.layers.GlobalAveragePooling2D()(base_model.output)
output = keras.layers.Dense(num_classes, activation="softmax")(avg)
model = keras.models.Model(inputs=base_model.input, outputs=output)
for layer in base_model.layers:
    layer.trainable = False
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.2,
    decay_steps=10000,
    decay_rate=0.01)
optimizer = keras.optimizers.SGD(learning_rate=lr_schedule, momentum=0.9)
model.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer,
              metrics=["accuracy"])
history = model.fit(train_ds,
                    steps_per_epoch=int(0.75 * len(train_ds) / batch_size),
                    validation_data=val_ds,
                    validation_steps=int(0.15 * len(train_ds) / batch_size),
                    epochs=5)
```

```python
# Xception fine tuning
for layer in base_model.layers:
    layer.trainable = True
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.01,
    decay_steps=10000,
    decay_rate=0.001)
optimizer = keras.optimizers.SGD(learning_rate=lr_schedule, momentum=0.9,
                                 nesterov=True)
model.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer,
              metrics=["accuracy"])
history = model.fit(train_ds,
                    steps_per_epoch=int(0.75 * len(train_ds) / batch_size),
                    validation_data=val_ds,
                    validation_steps=int(0.15 * len(train_ds) / batch_size),
                    epochs=40)
```

```python
predictions = model.predict(test_ds)
predictions = np.argmax(predictions,axis=1)
```
