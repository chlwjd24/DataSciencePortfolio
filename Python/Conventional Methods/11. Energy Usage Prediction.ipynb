## Linear Regression

```python
new_energy = energy.reset_index()

# create seconds from midnight column
def secSinceMidnight(datTimStr):
    tt = pd.to_datetime(datTimStr).time()
    return tt.hour * 3600 + tt.minute * 60 + tt.second

new_energy['NSM'] = new_energy.date.apply(secSinceMidnight)

# create week status column

def weekstatus(datTimStr):
    if pd.to_datetime(datTimStr).weekday() < 5:
        return "Weekday"
    return "Weekend"

new_energy['Weekstatus'] = new_energy.date.apply(weekstatus)

# create day of week column
new_energy['Day_of_week'] = new_energy['date'].dt.weekday_name
new_energy.head()
```

```python
new_energy['Weekday'] = (new_energy['Weekstatus'] == 'Weekday').astype('int64')
new_energy['Weekend'] = (new_energy['Weekstatus'] == 'Weekend').astype('int64')
new_energy['Monday'] = (new_energy['Day_of_week'] == 'Monday').astype('int64')
new_energy['Tuesday'] = (new_energy['Day_of_week'] == 'Tuesday').astype('int64')
new_energy['Wednesday'] = (new_energy['Day_of_week'] == 'Wednesday').astype('int64')
new_energy['Thursday'] = (new_energy['Day_of_week'] == 'Thursday').astype('int64')
new_energy['Friday'] = (new_energy['Day_of_week'] == 'Friday').astype('int64')
new_energy['Saturday'] = (new_energy['Day_of_week'] == 'Saturday').astype('int64')
new_energy['Sunday'] = (new_energy['Day_of_week'] == 'Sunday').astype('int64')
new_energy.head()
```

```python
feature_cols = np.array(['lights', 'T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4', 'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 
                'T8', 'RH_8', 'T9', 'RH_9', 'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed', 'Visibility', 'Tdewpoint',
                'NSM', 'Weekday', 'Weekend', 'Monday', 'Tuesday', 'Wednesday', 'Thursday',
                'Friday', 'Saturday', 'Sunday'])
X = new_energy[feature_cols]
y = new_energy['Appliances']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 142)

print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

The training shape contains 80% of the data 
The testing shape contains 20% of the data

Apply linear regression:

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
def mape(y_true, y_pred):
        return np.mean(np.abs((y_true - y_pred)/ y_true))*100
```

```python
y_train_pred = model.predict(X_train)
print("Training values:")
print("RMSE: ", math.sqrt(mean_squared_error(y_train, y_train_pred)))
print("R2: ", r2_score(y_train, y_train_pred))
print("MAE: ", mean_absolute_error(y_train, y_train_pred))
print("MAPE: ", mape(y_train, y_train_pred))
```

The training data from the report had an R-Squared of 0.18 which is the same value that we have.
The report had an RMSE score of 93.21 which is also very close to 92.77 in our model.
The MAE value is similar with 53.13 on the report and 52.71 in our model.
The MAPE values of 61.32% on the report is also similar to 60.72 from our model.
The results on the article compared to our results have almost no discrepancies.

```python
y_test_pred = model.predict(X_test)
print("Testing values:")
print("RMSE: ", math.sqrt(mean_squared_error(y_test, y_test_pred)))
print("R2: ", r2_score(y_test, y_test_pred))
print("MAE: ", mean_absolute_error(y_test, y_test_pred))
print("MAPE: ", mape(y_test, y_test_pred))
```

The report had an RMSE score of 93.18, which similar to our 94.92.
R-Sqaured was 0.16 from the report which is better han our 0.17 value.
MAE was 51.97 from the report, from our model we achieved 52.98.
The MAPE from the report 59.93, similar to 61.54 of our value.
Thus the results from the article were very simliar to what we achieved in our model.

## Recursive Feature Elimination

```python
X = new_energy[feature_cols]
y = new_energy['Appliances']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 142)

print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

Apply RFE:

```python
estimator = LinearRegression()
selector = RFE(estimator, 10) 
selector = selector.fit(X, y)

supp = selector.get_support()
print("Selected features:", feature_cols[supp])
print("Coeffs:", selector.estimator_.coef_)
# test the model
predicted = selector.predict(X)
```

```python
print("MSE:", mean_squared_error(y, predicted))
print("R^2:", r2_score(y, predicted))
```

In the model of recursive feature elimination, there were 10 features that were eliminated resulting in a mean squared error of
9059 and a R-squared of 0.14, which is quite low.
